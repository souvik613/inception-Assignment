#Q1:Temperature check

```python
current_temperature= float(input("Enter the current temperature:"))
if(current_temperature<10):
  print("it's too cold!")
elif(current_temperature>=10 and current_temperature<=25):
   print("it's cool outside")
else:
  print("it's too hot!")

```

#Q2:ATM withdrawal system

```python
accout_balance= float(input("Enter your account balance:"))
withdrawal_amount= float(input("Enter the amount you want to withdraw"))
if(withdrawal_amount>accout_balance):
  print("insufficient balance")
else:
  balance=accout_balance-withdrawal_amount;
  print("Withdrawal successful")
  print("Remaining balance:",balance)
```

#Q3:Student Grade calculator

```python
mark= int(input("Enter your mark: "))
if mark<=100 and mark >=90:
  print("A+")
elif(mark<=89 and mark>=80):
  print("A")
elif mark<=79 and mark >=70:
  print("B")
elif mark<=69 and mark>=60:
  print("c")
else:
  print("Fail")
```

#Q5:Restaurant Bill with discount


```python
total_bill= float(input("Enter the total bill amount:"))
if total_bill>=1000:
  discount=total_bill*0.10
  final_bill=total_bill-discount
else:
  final_bill=total_bill
print("Final bill amount:",final_bill)
```

#Leap year  checker

```python
year=int(input("Enter the year:"))
if(year%4==0):
  print("Leap year")
elif(year%4==0 and year%100!=0):
  print("Leap year")
else:
  print("Not a leap year")
```
